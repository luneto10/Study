```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np

sns.set_theme()
```

```python
df = pd.read_csv("../data/canadian_immegration_data.csv")
df.set_index("Country", inplace=True)
df.head()
```

```python
top_10 = df.sort_values(by="Total", ascending=False).head(10)
top_10
```

```python
sns.barplot(data=top_10, x=top_10.index, y="Total")
```

```python
ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h')
ax.set(xlabel='Total Imigrantes', ylabel='',
       title='10 Paises com mais Imigrantes para Canada')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h')

ax.set_title('10 Paises com mais Imigrantes para Canada', loc = 'left', fontsize = 16)
ax.set_xlabel('Total Imigrantes', fontsize = 14)
ax.set_ylabel('')
plt.show()
```

```python
def gerar_grafico_paleta(palette):
    fig, ax = plt.subplots(figsize = (10, 6))
    ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h', palette=palette, hue='Total')

    ax.set_title('10 Paises com mais Imigrantes para Canada', loc = 'left', fontsize = 16)
    ax.set_xlabel('Total Imigrantes', fontsize = 14)
    ax.set_ylabel('')
    plt.show()
```

```python
gerar_grafico_paleta('Blues')
```

```python
gerar_grafico_paleta('rocket_r')
```

```python
gerar_grafico_paleta('coolwarm')
```

```python
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style="dark")
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style="white")
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style="ticks")
gerar_grafico_paleta('tab10')
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h', palette='tab10', hue='Total', legend=False)

ax.set_title('10 Paises com mais Imigrantes para Canada', loc = 'left', fontsize = 16)
ax.set_xlabel('Total Imigrantes', fontsize = 14)
ax.set_ylabel('')
handles, labels = ax.get_legend_handles_labels()

# Reverse the order of handles and labels
handles.reverse()
labels.reverse()

# Set the legend with the reversed order
ax.legend(handles, labels, title='Total')

sns.despine()
plt.show()
```

```python
anos = list(map(str, range(1980, 2014)))
brasil = df.loc["Brazil", anos]
brasil_dict = {
    'ano' : brasil.index.tolist(),
    "imigrantes" : brasil.values.tolist()
}

dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```
