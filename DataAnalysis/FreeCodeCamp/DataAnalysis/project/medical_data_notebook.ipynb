```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../data/medical_examination.csv", index_col=0)
df.head()
```

```python
df[['cholesterol', 'gluc']] = df[['cholesterol', 'gluc']].replace({1 : 0})
df.head()
```

```python
df[['cholesterol', 'gluc']] = df[['cholesterol', 'gluc']].clip(upper=1)
df['overweight'] = 0
df.loc[df['weight'] / ((df['height'] / 100) ** 2) > 25, 'overweight'] = 1
df.head()
```

```python
df_melt = pd.melt(df, id_vars=['cardio'], value_vars=['active', 'alco', 'cholesterol', 'gluc', 'overweight', 'smoke'])
```

```python
df_counts = df_melt.groupby(['cardio', 'variable', 'value']).size().reset_index(name='total')
df_counts
```

```python
sns.catplot(
    x='variable', y='total', hue='value', col='cardio',
    data=df_counts, kind='bar', height=5, aspect=1.2
)
```
