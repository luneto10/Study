# More Visualizations

Previously, we saw an overview of how pandas `plot` method worked and how to use the basic API of matplotlib. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

### Global API 

```python
x = np.arange(-10,11)
```

```python
plt.figure(figsize=(12,6))
plt.title("My Nice Plot")

plt.plot(x, x**2)
plt.plot(x, -1 * (x**2))
```

### **OOP Interface**

```python
df = pd.read_csv("data/btc-market-price.csv", 
    header=None, 
    names=["Timestamp", "Price"], 
    index_col=0,
    parse_dates=True)

df.head()
```

```python
fig, axes = plt.subplots(figsize=(12,6))
axes.plot(df.index, df["Price"])
axes.set_ylabel("BTC Price")
```
